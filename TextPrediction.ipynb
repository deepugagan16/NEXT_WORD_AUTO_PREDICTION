# Readint the text from the unzipped file and then converting the text to *LOWERCASE*

```python
data=open('/content/data/next_word_predictor.txt','r').read().lower()
```

```python
data
```

# Creating a list of sentences present in the text using the *nltk.sent_tokenize()*

```python
sentences=sent_tokenize(data)
```

# *Tokenizing* the words in the text by assigning each word a key integer, so that computer understands the data

```python
tokenizer=Tokenizer()
tokenizer.fit_on_texts(sentences)
tokenizer.word_index
```

```python
print("Number of UNIQUE words in the dataset",len(tokenizer.word_index))
print("Number of sentences in the dataser",len(sentences))
```

# Converting each sentence to its *numerical representation*

```python
input_seq=[]
max_len=0 #to fint he max value for padding all the other sequences
for sentence in sentences:
  tokenized_sentence=tokenizer.texts_to_sequences([sentence])[0]
  max_len=max(max_len,len(tokenized_sentence))
  for i in range(1,len(tokenized_sentence)-1):
    input_seq.append(tokenized_sentence[:i+1])
```

```python
print("Longest sentene has",max_len,"words")
```

```python
for sentence in sentences:
  tokenized_sentence=tokenizer.texts_to_sequences([sentence])[0]
  if len(tokenized_sentence)==max_len:
    print(tokenizer.sequences_to_texts([tokenized_sentence])[0])
    break
```

# *Padding* the input sequences to make each sentence of the same length

```python
input_seq=pad_sequences(padding='pre',maxlen=max_len,sequences=input_seq)
```

```python
input_seq
```

# *Labelling* the data by separating the (INPUT, TARGET)

```python
X=input_seq[:,:-1]
Y=input_seq[:,-1]
Y=to_categorical(Y)
```

```python
X.shape
```

```python
Y.shape
```

# Creating the LSTM model (*with an Embedding layer in the front for better representation of the words*)

```python
# Explicitly define input_shape
model = Sequential()
model.add(Embedding(input_dim=len(tokenizer.word_index)+1,
                    output_dim=60,
                    input_length=max_len-1))  # input_length explicitly defined
model.add(LSTM(100))
model.add(Dense(len(tokenizer.word_index)+1, activation='softmax'))
model.build(input_shape=(None, max_len-1))
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
model.fit(X,Y,epochs=60)
```
